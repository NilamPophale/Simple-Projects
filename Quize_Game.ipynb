## Python Quize game

```python
questions=("What is the capital of India?",
          "How many bones in the human body?",
          "How many minutes are in a full week?",
          "How many dots appear on a pair of dice?",
          "How many hearts does an octopus have?",
          "How many colors are used in the South African flag? ",
          "Where did sushi originate?",
          "What company was originally called 'Cadabra'?",
          "How many elements are in the periodic table?",
          "Which animal lays the largest eggs?")

options= (('A.Mumbai','B.Delhi','C.Pune','D.Nagar'),
          ('A.205','B.206','C.207','D.208'),
          ('A.10000','B.10050','C.10080','D.10090'),
          ('A.42','B.43','C.40','D.41'),
          ('A.1','B.2','C.3','D.4'),
          ('A.3','B.4','C.5','D.6'),
          ('A.Canada','B.China','C.Japan','D.Korea'),
          ('A.Ajio','B.Meesho','C.Amazon','D.Flipkart'),
          ('A.117','B.118','C.119','D.120'),
          ('A.Whale','B.Crocodile','C.Elephant','D.Ostrich'))

answers=('B','B','C','A','C','D','B','C','B','D')
guesses=[]
score=0
question_num=0

for question in questions:
    print('----------------------------------------------')
    print(question)
    for option in options[question_num]:
        print(option)
        
    guess = input('Enter (A, B, C, D):').upper()
    guesses.append(guess)
    if guess==answers[question_num]:
        score+=1
        print('correct!')
    else:
        print('Incorrect!')
        print(f'{answers[question_num]}is the correct answer')
    question_num+=1
        
print('---------------------------------')
print('              Results            ')
print('---------------------------------')

print('answers: ', end='')
for answer in answers:
    print(answer,end='')
    
print()

print('guesses: ', end='')
for guess in guesses:
    print(guess,end='')
    
print()

score = int(score / len(questions) * 100)
print(f'Your scores is: {score}%')
```
